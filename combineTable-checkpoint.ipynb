```python
import plotly.graph_objects as go
import ipywidgets as widgets
from datetime import date, timedelta
from IPython.display import display, clear_output, HTML, display_html, IFrame
import numpy as np
pd.set_option('display.max_rows', None)  # 设置为 None 表示无限制
pd.set_option('display.max_columns', None)  # 设置为 None 表示无限制
```

```python
# 定义一个函数，将数值格式化为字符串，根据不同的列应用不同的格式
def format_column(col_name, series):
    if col_name in ['销量', '展示量', '点击量', '广告订单', '广告销量', '自然销量']:
        return series.astype(int).astype(str)
    elif col_name in ['销售额', '折扣', '广告费', '广告销售额', 'CPC', 'CPO']:
        return series.apply(lambda x: f'${x:.1f}')
    elif col_name in ['广销比', '折扣比', 'tAcos', '点击率', 'ad转化率', 'Acos']:
        return series.apply(lambda x: f'{x:.1%}')
    else:
        return series

# 更新 display_scrollable_df 函数以应用数据格式化
def display_scrollable_df(df, height='500px'):
    # 格式化数据
    for col in df.columns:
        df[col] = format_column(col, df[col])
    
    # 将DataFrame转换为HTML表格字符串
    df_html = df.to_html(index=False, escape=False)  # 设置 escape=False 以正确显示格式化的字符串
    
    # 使用HTML和CSS创建固定表头和滚动体的表格
    html = f'''
    <html>
    <head>
        <style>
            .scrollable {{
                height: {height};
                overflow-y: auto;
                margin: 0;
            }}
            .dataframe {{
                border-collapse: collapse;
                width: 100%;
            }}
            .dataframe th, .dataframe td {{
                border: 1px solid black;
                padding: 8px;
                text-align: left;
                font-size: 10px;  # 调整字体大小
                min-width: 200px;  # 设置最小列宽
            }}
            .dataframe thead th {{
                position: sticky;
                top: 0;
                background-color: #f8f8f8;
            }}
        </style>
    </head>
    <body>
        <div class="scrollable">
            {df_html}
        </div>
    </body>
    </html>
    '''
    display_html(html, raw=True)

#绘图变量选择器
column_selector = widgets.SelectMultiple(
    options=["销售额", "广销比", "折扣比", "tAcos", "点击率", "CPC", "ad转化率", "自然销量", "Acos", "CPO"],
    value=['销售额', '广销比'],  # 默认选中的列
    description='选择变量',
    disabled=False,
    style={'height': '400px'}  # 调整高度以显示所有选项
)

#创建结束日期选择器
min_date = combine_table["日期"].min()
max_date = combine_table["日期"].max()
end_date_picker = widgets.DatePicker(
    description = "结束日期", 
    disabled = False, 
    value = max_date,  #默认为数据集最新的日期
    min = min_date, 
    max = max_date)
#创建开始日期选择器
start_date_picker = widgets.DatePicker(
    description = "开始日期", 
    disabled = False, 
    value = end_date_picker.value - timedelta(days = 7),  #默认开始时间为7天前
    min = min_date, 
    max = max_date)

#创建日期聚合方式选择器
date_aggregate_option = widgets.ToggleButtons(
    options = ["按天", "按周", "按月"],
    description = "视图",
    disabled = False, 
    button_style = ""  # 样式'success', 'info', 'warning', 'danger' or ''这里选默认
)

#创建负责人选择器
responsible_option = widgets.Dropdown(
    options = [""] + sorted(combine_table["负责人"].unique().tolist()), 
    description = "负责人",
    value = "", 
    disabled = False
)
#创建类型选择器
type_option = widgets.Dropdown(
    options= [""] + sorted(combine_table['类型'].unique().tolist()),
    description='类型:',
    value = "",
    disabled=False,
)
# 组合选择器
combo_option = widgets.Dropdown(
    options=[""] + sorted(combine_table['组合'].unique().tolist()),
    description='组合:',
    value = "", 
    disabled=False,
)

# 创建一个新的输出部件用于绘图
plot_output = widgets.Output()

#绘图函数
def draw_plot(data, x, ys):
    with plot_output:
        plot_output.clear_output(wait=True)  # 在重新绘图前清除现有图形
        
        # 创建一个Plotly图形对象
        fig = go.Figure()

        # 计算y轴位置偏移
        axis_count = len(ys)
        axis_spacing = 1 / (axis_count + 1)
        
        # 为每个y轴变量添加折线图层
        for i, y in enumerate(ys):
            # 添加轨迹
            fig.add_trace(go.Scatter(
                x=data[x], 
                y=data[y], 
                mode='lines+markers', 
                name=y,
                yaxis=f'y{i + 1}'  # 每个变量使用不同的y轴
            ))

            # 动态添加y轴到图形布局
            fig.update_layout(**{
                f'yaxis{i + 1}': {
                    'title': y,
                    'side': 'left' if i % 2 == 0 else 'right',
                    'overlaying': 'y' if i > 0 else None,
                    'position': 0.05 * i if i % 2 == 0 else 1 - 0.05 * i
                }
            })

        # 设置图表的标题和x轴标签
        fig.update_layout(
            title="数据趋势分析",
            xaxis_title=x,
            hovermode="x unified",
            width=1000,
            height=400
        )

        # 确保图表目录存在
        if not os.path.exists('plotly_plots'):
            os.mkdir('plotly_plots')
        filepath = 'plotly_plots/plot.html'
        fig.write_html(filepath)

        # 显示图表，使用 IFrame
        display(IFrame(src=filepath, width=1100, height=410))
        
# 创建了一个输出表格的小部件。这个小部件可以捕获和显示来自 Python 代码的输出
output = widgets.Output()

#生成表格函数
def filter_data (*args): #选择器值变化更新表格函数*args接受任意数量的位置形参
    #使用上下文管理器with output表示接下来的代码块将会在 output小部件的上下文中执行。
    #这意味着该代码块中生成的所有输出都将直接显示在这个 output 小部件中
    #而不是在标准的输出区域（如终端或Jupyter的输出单元）。
    with output:
        #清除 output 小部件中的当前内容。如果 wait 参数设置为 True，它将等待直到新的输出被生成才进行清除。
        output.clear_output(wait=True) 
        #按照日期选择器部件的值变化更新表格
        start_date = pd.Timestamp(start_date_picker.value) #Timestamp类似pd.to_datetime()
        end_date = pd.Timestamp(end_date_picker.value) #由于 DatePicker 控件返回的是 datetime.date需要转换日期格式才能进行比较
        if end_date < start_date:
            end_date_picker.value = start_date
            return
        #筛选符合日期范围的数据
        filtered_data = combine_table[(combine_table["日期"] >= start_date) & (combine_table["日期"] <= end_date)
        ]

        # 应用基于选择器的筛选条件
        if responsible_option.value:
            filtered_data = filtered_data[filtered_data['负责人'] == responsible_option.value]
        if type_option.value:
            filtered_data = filtered_data[filtered_data['类型'] == type_option.value]
        if combo_option.value:
            filtered_data = filtered_data[filtered_data['组合'] == combo_option.value]
        
        #基于选择器选择，分组键列表
        group_keys = ["日期"]
        if responsible_option.value:
            group_keys.append('负责人')
        if type_option.value:
            group_keys.append('类型')
        if combo_option.value:
            group_keys.append('组合')
        
        #按日期聚合方式选择器更新表格
        numeric_columns = filtered_data.select_dtypes(include = ["number"]).columns.tolist()  #提取数值类型的列名
        if date_aggregate_option.value == "按天":
            filtered_data = filtered_data.groupby(group_keys, observed=True)[numeric_columns].sum().reset_index()
        elif date_aggregate_option.value == "按周":
            #.dt.to_period('W'): 将datetime对象转换为按周时间周期。
            #.apply(lambda r: r.start_time): 获取每个周周期的开始日期。
            filtered_data["日期"] = pd.to_datetime(filtered_data["日期"].dt.to_period("W").apply(lambda r: r.start_time))
            filtered_data = filtered_data.groupby(group_keys, observed=True)[numeric_columns].sum().reset_index()
        elif date_aggregate_option.value == "按月":
            filtered_data["日期"] = pd.to_datetime(filtered_data["日期"].dt.to_period("M").apply(lambda r: r.start_time))
            filtered_data = filtered_data.groupby(group_keys, observed=True)[numeric_columns].sum().reset_index()

        # 计算广销比
        filtered_data['广销比'] = np.where(filtered_data['销售额'] != 0, filtered_data['广告费'] / filtered_data['销售额'], np.nan)
        # 计算折扣比
        filtered_data['折扣比'] = np.where(filtered_data['销售额'] != 0, filtered_data['折扣'] / filtered_data['销售额'], np.nan)
        # 计算 tAcos
        filtered_data['tAcos'] = np.where(filtered_data['销售额'] != 0, (filtered_data['广告费'] + filtered_data['折扣']) / filtered_data['销售额'], np.nan)
        # 计算点击率
        filtered_data['点击率'] = np.where(filtered_data['展示量'] != 0, filtered_data['点击量'] / filtered_data['展示量'], np.nan)
        # 计算 CPC
        filtered_data['CPC'] = np.where(filtered_data['点击量'] != 0, filtered_data['广告费'] / filtered_data['点击量'], np.nan)
        # 计算转化率
        filtered_data['ad转化率'] = np.where(filtered_data['点击量'] != 0, filtered_data['广告销量'] / filtered_data['点击量'], np.nan)
        # 计算自然销量
        filtered_data['自然销量'] = filtered_data['销量'] - filtered_data['广告销量']
        # 计算 Acos
        filtered_data['Acos'] = np.where(filtered_data['广告销售额'] != 0, filtered_data['广告费'] / filtered_data['广告销售额'], np.nan)
        # 计算 CPO
        filtered_data['CPO'] = np.where(filtered_data['广告销量'] != 0, filtered_data['广告费'] / filtered_data['广告销量'], np.nan)
   
        #复制表格
        summary_data = filtered_data.copy()
        summary_data["日期"] = summary_data["日期"].dt.date
        
        # 计算汇总行前先进行必要的总和计算
        total_ad_cost = filtered_data['广告费'].sum()
        total_sales = filtered_data['销售额'].sum()
        total_discount = filtered_data['折扣'].sum()
        total_clicks = filtered_data['点击量'].sum()
        total_impressions = filtered_data['展示量'].sum()
        total_ad_sales = filtered_data['广告销售额'].sum()
        total_ad_sales_volume = filtered_data['广告销量'].sum()
        # 计算汇总指标
        summary = {
            "日期" : "汇总", 
            "销量" : filtered_data["销量"].sum(),
            "销售额" : filtered_data["销售额"].sum(),
            "折扣" : filtered_data["折扣"].sum(),
            "展示量" : filtered_data["展示量"].sum(),
            "点击量" : filtered_data["点击量"].sum(),
            "广告费" : filtered_data["广告费"].sum(),
            "广告销售额" : filtered_data["广告销售额"].sum(),
            "广告订单" : filtered_data["广告订单"].sum(),
            "广告销量" : filtered_data["广告销量"].sum(),
            '广销比': total_ad_cost / total_sales if total_sales != 0 else np.nan,
            '折扣比': total_discount / total_sales if total_sales != 0 else np.nan,
            'tAcos': (total_ad_cost + total_discount) / total_sales if total_sales != 0 else np.nan,
            '点击率': total_clicks / total_impressions if total_impressions != 0 else np.nan,
            'CPC': total_ad_cost / total_clicks if total_clicks != 0 else np.nan,
            'ad转化率': total_ad_sales_volume / total_clicks if total_clicks != 0 else np.nan,
            '自然销量': total_sales - total_ad_sales,
            'Acos': total_ad_cost / total_ad_sales if total_ad_sales != 0 else np.nan,
            'CPO': total_ad_cost / total_ad_sales_volume if total_ad_sales_volume != 0 else np.nan
        }
        # 将汇总行添加到 DataFrame 中
        summary_df = pd.DataFrame([summary], index=['汇总'])
        summary_data_table = pd.concat([summary_data, summary_df])
        
        #显示表格
        display_scrollable_df(summary_data_table)
        #display(filtered_data) # 使用display函数，以便在Jupyter中更好地显示DataFrame
        
        # 选定绘图的列
        selected_columns = column_selector.value  # 示例：根据用户互动动态选择这些列
        draw_plot(filtered_data, '日期', selected_columns)
    
#为选择器部件添加观察者，将filter_data函数绑定到日期选择器的变化事件，使得每当日期选择器的值改变时，filter_data函数自动再次运行。
column_selector.observe(filter_data, "value")
start_date_picker.observe(filter_data, "value")
end_date_picker.observe(filter_data, "value")
date_aggregate_option.observe(filter_data, "value")
responsible_option.observe(filter_data, "value")
type_option.observe(filter_data, "value")
combo_option.observe(filter_data, "value")

#设置初始化用户界面布局，一开始是空的
# 绘图变量选择器
display(column_selector)
display(plot_output)
display(start_date_picker, end_date_picker)
display(date_aggregate_option)
display(responsible_option)
display(type_option)
display(combo_option)
display(output)
#调用函数根据当前选择更新显示内容
filter_data()
```
